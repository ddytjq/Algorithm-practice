```python
# stop watch

from tkinter import *

root=Tk()
root.geometry("300x300")
root.title("stop watch")
var=0
change=False
def func1():
    global change
    change=True
def func2():
    global change
    change=False
def func():
    global var
    if change==True:
        var+=1
        label.config(text="%d" % int(var))
    root.after(100, func)
    
label=Label(root, text="0", font=(" ", 60))
button1=Button(root, text="bt1", command=func1, width=4, height=2)
button2=Button(root, text="bt2", command=func2, width=4, height=2)
label.pack()
button1.pack()
button2.pack()

func()

root.mainloop()
```

```python
# 숫자 맞추기

from tkinter import *
from random import *

root=Tk()
root.geometry("500x200")
root.title("game")
num=randint(1, 100)
print("정답 : %d" % num)

def func1():
    
    if int(entry.get()) == num:
        label2.config(text="정답!")
    elif int(entry.get()) > num:
        label2.config(text="높음!")
    elif int(entry.get()) < num:
        label2.config(text="낮음!")

def func2():
    global num
    num=randint(1, 100)
    label2.config(text="정답 초기화!")
    print("정답 : %d" % num)
    
label1=Label(root, text="숫자를 맞춰보세용")
entry=Entry(root)
bt1=Button(root, text="클릭!", command=func1)
bt2=Button(root, text="초기화!", command=func2)
label2=Label(root, text="정답판")

label1.place(x=180, y=10)
entry.place(x=10, y=40)
bt1.place(x=203, y=43)
bt2.place(x=237, y=43)
label2.place(x=287, y=43)

root.mainloop()
```

```python
# gui 연습 1

from tkinter import *

root=Tk()
root.geometry("400x200")
root.title("몸무게")
var=IntVar()
var.set(0)
num=0
def func1():
    global num
    num=50
def func2():
    global num
    num=53
def func3():
    global num
    num=58
def func4():
    global num
    num=67
def func5():
    global num
    num=75

def func():
    if num == 50:
        la2.config(text="1")
    elif num == 53:
        la2.config(text="2")
    elif num == 58:
        la2.config(text="3")
    elif num == 67:
        la2.config(text="4")
    elif num == 75:
        la2.config(text="5")
    
la1=Label(root, text="몸무게를 맞춰라!!!!", font=(" ", 15))
rb1=Radiobutton(root, text="1번 : 50kg", variable=var, value=1, command=func1)
rb2=Radiobutton(root, text="2번 : 53kg", variable=var, value=2, command=func2)
rb3=Radiobutton(root, text="3번 : 58kg", variable=var, value=3, command=func3)
rb4=Radiobutton(root, text="4번 : 67kg", variable=var, value=4, command=func4)
rb5=Radiobutton(root, text="5번 : 75kg", variable=var, value=5, command=func5)
bt1=Button(root, text="정답 버튼", command=func)
la2=Label(root, text=" ", font=(" ", 20))
la1.place(x=125, y=10)
rb1.place(x=10, y=30)
rb2.place(x=10, y=50)
rb3.place(x=10, y=70)
rb4.place(x=10, y=90)
rb5.place(x=10, y=110)
bt1.place(x=10, y=140)
la2.place(x=200, y=70)

root.mainloop()
```

```python
#gui 연습 2

from tkinter import *

root=Tk()
root.geometry("400x200")
root.title("몸무게")
var=IntVar()
var.set(0)
num=0
def func1(ev):
    la2.config(text="몸무게 : %d" % int(ev))
    global num
    num = int(ev)
    
def func():
    if num >=60 and num <90:
        la2.config(text="정답!!!")
    else:
        la2.config(text="오답!!!")
        
la1=Label(root, text="몸무게를 맞춰라!!!!", font=(" ", 15))
scale=Scale(root, from_=0, to=100, variable=var, showvalue='False', orient='h', label='몸무게', command=func1)
bt1=Button(root, text="정답 버튼", command=func)
la2=Label(root, text="몸무게 : ", font=(" ", 20))

la1.place(x=125, y=10)
scale.place(x=30, y=40)
bt1.place(x=30, y=85)
la2.place(x=180, y=50)

root.mainloop()
```
