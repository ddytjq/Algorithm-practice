```python
# 라디오버튼

from tkinter import *
# mesagebox.showinfo -> 메세지박스 출력
from tkinter import messagebox

root=Tk()
a=IntVar()
a.set(1)

def func():
    if a.get() == 1:
        str="Radio 1 selected"
    elif a.get() == 2:
        str="Radio 2 selected"
    else:
        str="아무것도 선택되지 않았습니다."
    messagebox.showinfo("샷",str)

button=Button(root, text="선택", command=func)
rd1=Radiobutton(root, text="남자", variable=a, value=1)
rd2=Radiobutton(root, text="여자", variable=a, value=2)

rd1.pack()
rd2.pack()
button.pack()

root.mainloop()
```

```python
# 스톱워치 만들기

from tkinter import *

root=Tk()
running = False # 시작은 false
timer=0

def function():
    global timer
    if (running==True) :
        timer+=1
        label.config(text=str(timer))
    
    # 100ms마다 계속 반복
    root.after(100, function)

def button1():
    global running
    running = True
    
def button2():
    global running
    running = False
    
label=Label(root, text='0')
button1=Button(root, text='시작', command=button1)
button2=Button(root, text='중지', command=button2)

label.pack()
button1.pack()
button2.pack()

function()

root.mainloop()
```
